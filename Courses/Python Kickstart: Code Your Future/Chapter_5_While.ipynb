- In this chapter, we will will cover `while` loop. A `while` loop in Python repeatedly executes a target statement as long as a given condition is true.
- It is used when the number of iterations is not known beforehand.

## Syntax
The basic syntax of a `while` loop is:
```python
while condition:
    # code block to be executed
```

### 1. Basic While Loop
- A simple example of a `while` loop that prints numbers from 1 to 5.

````python
# This print the number from 1 to 5
i = 1 
while i <= 5: # The loops is enabled if i<= 5
    print(f"Prints {i} because {i} is less or equal to 5.")
    i += 1 # Adds 1 to i variable
````

### 2. Using a Break Statement
- The `break` statement can be used to exit the loop when a certain condition is met.

````python
# This code execute a while loop with break

i = 1
while i <= 5: # if i is less or equal to 5, the loop is executed
    if i == 3: # if i is equal to 3 the while loop is ended.
        print(f"Prints i = {i} and the while loop ends")
        break
    print(f"Prints i because {i} less or equal to 5 and {i} is not equal to 3. Loops continue to the next iteration.")
    i += 1
````

### 3. Using a Continue Statement
- The `continue` statement can be used to skip the current iteration and continue with the next.

````python
# This code execute a while loop with continue

i = 1
while i <= 5: # if i is less or equal to 5, the loop is executed
    i += 1
    if i == 3: # if i is equal to 3 the atual iteration is ended and all the code below is not executed.
        print(f"Prints i = {i} and the iteration ends. All the text below continue is not executed.")
        continue
    print(f"Prints i because {i} less or equal to 5 and {i} is not equal to 3. Loops continue to the next iteration.")
````

### 4. While Loop with Else
- The `else` block will execute when the condition becomes false.

````python
# This print the number from 1 to 5

i = 1 
while i <= 5: # The loops is enabled if i<= 5
    print(f"Prints {i} because {i} is less or equal to 5.")
    i += 1
else:
    print('Loop is over')
````

### 5. Infinite Loops
